
## Navigating_Future_Online_Shopping

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
customers = pd.read_csv('/content/sample_data/customers.csv')
orders = pd.read_csv('/content/sample_data/orders.csv')
order_items = pd.read_csv('/content/sample_data/order_items.csv')
payments = pd.read_csv('/content/sample_data/payments.csv')
products = pd.read_csv('/content/sample_data/products.csv')
sellers = pd.read_csv('/content/sample_data/sellers.csv')
geolocation = pd.read_csv('/content/sample_data/geolocation.csv')
```

```python
for name, df in {
    'customers': customers,
    'orders': orders,
    'order_items': order_items,
    'payments': payments,
    'products': products,
    'sellers': sellers,
    'geolocation': geolocation
}.items():
    print(f'\n{name.upper()}')
    print(df.shape)
    display(df.head())
```

```python
customers.columns = [
    'customer_id',
    'customer_unique_id',
    'customer_zip_code_prefix',
    'customer_city',
    'customer_state'
]
customers.head()
```

```python
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])
```

```python
sales = orders.merge(order_items, on='order_id', how='inner')
total_orders = orders['order_id'].nunique()
total_revenue = sales['price'].sum()
avg_order_value = total_revenue / total_orders
total_orders, total_revenue, avg_order_value
```

##Data Visualization

```python
sales['month'] = sales['order_purchase_timestamp'].dt.to_period('M')
monthly_revenue = sales.groupby('month')['price'].sum()
monthly_revenue.plot(kind='line', title='Monthly Revenue Trend')
plt.show()
```

```python
customers['customer_state'].value_counts().plot(kind='bar')
plt.title('Customers by State')
plt.show()
```

```python
sales_prod = sales.merge(products, on='product_id', how='left')
sales_prod.rename(columns={'product category': 'product_category'}, inplace=True)
category_revenue = sales_prod.groupby('product_category')['price'].sum().sort_values(ascending=False)
category_revenue.head(10).plot(kind='bar', title='Top Product Categories')
plt.show()
```

```python
payments['payment_type'].value_counts().plot(kind='bar', title='Payment Methods')
plt.show()
```

```python
sales.groupby(sales['order_purchase_timestamp'].dt.year)['price'].sum().plot(kind='bar', title='Yearly Revenue')
plt.show()
```

```python
daily_sales = orders.merge(order_items, on='order_id')
daily_sales['date'] = daily_sales['order_purchase_timestamp'].dt.date
daily_sales.groupby('date')['price'].sum().cumsum().plot()
plt.title('Cumulative Sales')
plt.show()
```
